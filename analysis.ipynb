```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("crimes_against_women_2001-2014.csv")

data.head()
```

```python
data.columns.tolist()
```

```python
len(data)
```

```python
data = data.drop("Unnamed: 0",axis = 1)
```

```python
# check for missing/null values
missing_values = data.isna().sum()
missing_values
```

```python
data.drop_duplicates(inplace = True)
```

```python
data
```

```python
for i,j in zip(data['STATE/UT'],data['DISTRICT']):
    if i == 'Puducherry' or i == 'Lakshadweep':
        print(j)
```

```python
data['STATE/UT'] = data['STATE/UT'].str.upper()
```

```python
data.head()
```

```python
data['STATE/UT'].unique()
```

```python
data['STATE/UT'] = data['STATE/UT'].replace('DELHI UT','DELHI')
data['STATE/UT'].unique()
```

```python
data['STATE/UT'] = data['STATE/UT'].replace('A & N ISLANDS','A&N ISLANDS')
data['STATE/UT'] = data['STATE/UT'].replace('D & N HAVELI','D&N HAVELI')
```

```python
data['STATE/UT'].unique()
```

```python
data['STATE/UT'] = data['STATE/UT'].replace('TELANGANA','ANDHRA PRADESH')
data['STATE/UT'].unique()
```

```python
for i in data['DISTRICT']:
    if 'Total' in i or "TOTAL" in i:
        data['DISTRICT'] = data['DISTRICT'].replace(i,'TOTAL')
```

```python
data.head()
```

```python
# 2001 analysis
dict_2001 = {}
maxi = 0
max_states = {}
# Iterate over each row in the DataFrame
for i in range(len(data)):
    # Check if the year is 2001 and the district is "TOTAL"
    if data.iloc[i]['Year'] == 2001 and data.iloc[i]['DISTRICT'] == "TOTAL":
        # Calculate the sum of the specified columns
        sum_crimes = data.iloc[i][['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']].sum()
        
        # Store the result in the dictionary
        dict_2001[data.iloc[i]['STATE/UT']] = sum_crimes

        if maxi < sum_crimes:
            maxi = sum_crimes
            max_state = data.iloc[i]['STATE/UT']

max_states[f"{max_state} 01"] = maxi
crime_data = dict_2001

filtered_crime_data = {state: value for state, value in crime_data.items() if value > 2000}

# Sort the filtered data by values in descending order
filtered_crime_data = dict(sorted(filtered_crime_data.items(), key=lambda item: item[1], reverse=True))

# Plotting the bar graph
plt.figure(figsize=(15, 8))
plt.bar(filtered_crime_data.keys(), filtered_crime_data.values(), color='skyblue')
plt.xlabel('State/UT')
plt.ylabel('Number of Cases')
plt.title('States/UTs with Number of Cases Greater than 2000 in 2001')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit everything

# Display the plot
plt.show()

year = "2001"
crime_sum = 0
for i in crime_data:
    crime_sum += crime_data[i]

total_crime_data = {}

total_crime_data[year] = crime_sum/365
```
